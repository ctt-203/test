```python
class Dog:
    def __init__(self, name):
        self.name = name
        self.tricks = []
    def add_trick(self, trick):
        self.tricks.append(trick)
    def __repr__(self):
        return self.name
```

```python
d1 = Dog("doge")
d2 = Dog("hajimi")
d1.add_trick("haren")
```

```python
print(d1)
```

```python
print(d2.tricks)
```

```python
class ListNode:
    def __init__(self, value, next_):
        self.value = value
        self.next_ = next_
```

```python
l = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5, None)))))
```

```python
import lolviz
```

```python
lolviz.objviz(l)
```

```python
count = 0
n = l
while n != None:
    count += n.value
    n = n.next_
print(count)
```

```python
def sum_list(l):
    if l is None:
        return 0
    else:
        return l.value + sum_list(l.next_)
```

```python
sum_list(l)
```

```python
def reverse(l):
    if l is None:
        return None, None
    elif l.next_ is None:
        ret = ListNode(l.value, None)
        return ret, ret
    else:
        a, b = reverse(l.next_)
        tail = ListNode(l.value, None)
        b.next_ = tail
        return a, tail
```

```python
m = reverse(l)
```

```python
lolviz.objviz(m)
```
